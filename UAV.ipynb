```python
import numpy as np
import matplotlib.pyplot as plt

diameter = np.array([3,5,6,10,14,15])
pitch = np.array([1.5,3.2,4,6,4.5,5])
thrust = 2.107
r = np.sqrt(thrust/(18.59*(10**-12)*(diameter**3.5)*(pitch**0.5)))
W = 5.3*(10**(-15))*(r**3)*(diameter**4)*(pitch)
current=[]
voltage = []
kv = np.arange(100,1600,100)
fig, axs = plt.subplots(2, 1,sharex = True)

for i in range(0,len(W)):
    
    
        I = kv*(W[i]/r[i])
        v = r[i]/kv
        
        axs[0].plot(kv,v,label = "d =%d"%(diameter[i]))
        axs[0].set_title('Kv vs voltage')
        axs[0].set_xlabel('kv')
        axs[0].set_ylabel('voltage')
        fig.suptitle('Kv vs voltage and Kv vs current', fontsize=16)

        axs[1].plot(kv,I, label = "d =%d"%(diameter[i]))
        axs[1].set_xlabel('Kv')
        axs[1].set_title('Kv vs Current')
        axs[1].set_ylabel('Current')
plt.legend()
plt.show()
```
